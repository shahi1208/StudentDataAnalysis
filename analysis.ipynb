```python
import pandas as pd
```

```python
data = pd.read_csv("studentdata.csv")
data.head()
```

```python
data.columns
```

```python
data[' STUDENT_BRANCH'] = data[' STUDENT_BRANCH'].replace('ECONOMICS','ENGLISH')
data[' STUDENT_BRANCH'].unique()
```

```python
data['COURSE']
```

```python
data['COURSE'] = data['COURSE'].replace(['MATHS-BIOLOGY','BIOLOGY/MATHS','BIO /  MAT','BIOLOGY-MATHEMATICS', 'BIOLOGY-MATHS','BIOLOGY WITH MATHS','BIO/MAT','BIO-MATHS', 'bio+math','BIOLOGY- MATHEMATICS','BIOLOGY/MATHEMATICS'],'BIO+MATHS')
```

```python
data['COURSE'] = data['COURSE'].replace(['COMPUTER SCIENCE/MATHEMATICS','MATHS - COMPUTER SCIENCE', 'COMPUTER SCIENCE-MATHS','COMPUTER SCIENCE -MATHS', 'CS / MAT','CS/MAT','MATHS- COMPUTER SCIENCE','MATHS/COMPUTER SCIENCE','MATHS-COMPUTER SCIENCE', 'MATHS -COMPUTER SCIENCE', 'CS-MATHS', 'MAT/CS','CS/MAT','MATHS - COMPUTER SCIENCE','COMPUTER SCIENCE'] , 'COMP SCI+MATHS')
```

```python
data['COURSE'] = data['COURSE'].replace(['COMMERCE/ACCOUNTANCY','ACC / CS','COMMERCE / COMPUTER SCIENCE','COMPUTER SCIENCE-COMMERCE', 'COMMERCE - COMPUTER SCIENCE', 'COMMERCE COMPUTER SCIENCE' 'COMMERCE- COMPUTER SCIENCE','COMMERCE-COMPUTER SCIENCE', 'ACCOUNTANCY-COMPUTRE SCIENCE','ACCOUNTANCY-COMPUTER', 'COMMERCE', 'ACCOUNTS-COMPUTER SCIENCE','COMMERCE/COMPUTER SCIENCE','ACCOUNTS - COMPUTER SCIENCE', 'MATHS - COMPUTER SCIENCE','ACCOUNTS/COMPUTER SCIENCE','commerce', 'COMMERCE COMPUTER SCIENCE', 'COMMERCE- COMPUTER SCIENCE','COMMERCE-ACCOUNTANCY', 'COMPUTER SCIENCE WITH ACCOUNTANCY', 'COMPUTER WITH ACCOUNTANCY', 'ACCOUNTANCY-COMPUTER SCIENCE'], 'COMMERCE+COMP')
```

```python
data['COURSE'] = data['COURSE'].replace(['ACC/BUS MAT','ACCOUNTANCY/BUSINESS MATHS', 'ACC/BUS MATHS', 'ACCOUNTS-BUSINESS MATHS', 'ACCOUNTS- BUSINESS MATHS','ACC/BUSINESS MATHS', 'BUSINESS STUDIES', 'COMMERCE-BUSINESS MATHS', 'ACCOUNTS-BUSINESS STUDIES', 'BUSINESS MATHS-COMMERCE','ACCOUTNS-BUSINESS MATHS','COMMERCE -BUSINESS MATHS', 'ACCOUNTS - BUSINESS MATHS','COMMERCE/BUSINESS MATHS','BUSINESS-COMMERCE'], 'ACCS+BUS MATH')
```

```python
data['COURSE'] = data['COURSE'].replace(['BIOLOGY','PURE SCIENCE','BOTANY-ZOOLOGY','BIOLOGY-HOME SCIENCE','BIOLOGY  ','SCIENCE','BOTANY/ZOOLOGY','BIOLOGY/NUTRITION & DIETETICS','PURESCIENCE'], 'PURE SCIENCE')
data['COURSE'] = data['COURSE'].replace(['ACCOUNTS-COMMUNICATIVE ENGLISH','BMAT', 'ACCOUNTS-HISTORY', 'ACCOUNTS-TYPEWRITING','nan','HISTORY','ACCOUNTS/HISTROY'], 'OTHERS')
data['COURSE']= data['COURSE'].replace(['BIO/CS','BIOLOGY-COMPUTER SCIENCE','BIOLOGY/COMPUTER SCIENCE'],'BIO+COMP')
data['COURSE']
```

```python
data['RELIGION'] = data['RELIGION'].replace(['HINDU','hindu','hindu vanniyar','Hindu','HINHU','HINDI'], 'Hindu')
data['RELIGION'] = data['RELIGION'].replace(['MUSLIM', 'muslim', 'ISLAM'], 'Muslim')
data['RELIGION'] = data['RELIGION'].replace(['CHRISTIAN', 'christian','ROMAN CATHOLIC', 'ANGLO INDIAN', 'CATHOLIC','CHIRISTIAN', 'R.C.CHRISTIAN', 'Christian', 'CHRISTIANITY'], 'Christian')
data['RELIGION']
```

```python
data['COMMUNITY'].replace('BCM','BC-M', inplace= True)
data['COMMUNITY'].replace('BCC','BC-C', inplace= True)
data['COMMUNITY'].replace(['BCA','BC(A)'],'BC-A', inplace= True)
data['COMMUNITY'].replace('Sc','SC', inplace= True)
data['COMMUNITY']
```

```python
#data.columns
data['BACKGROUND'].replace(['MATHUR','Semi - Urban'], 'Semi-urban',inplace= True)
data['BACKGROUND'].replace(['urban','URBAN'],'Urban', inplace=True)
data['BACKGROUND'].unique()

data['BOARD'].replace(['stateboard','state board','HSC','STATE BOAD', 'STATE BOARS'], 'STATE BOARD', inplace= True)
data['BOARD'].replace(['a mole on the neck', '-','BLANK'], '0', inplace= True)
data['BOARD'].replace(['CENTRAL BOARD', 'CBSC'], 'CBSE', inplace= True)
data['BOARD'].replace(' icse board ','ICSE', inplace= True)
data
```
